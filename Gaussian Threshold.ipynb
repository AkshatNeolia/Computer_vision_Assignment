**Objective** - Classical filtering option or Threshold techniques on image dataset

**Classical Filtering:**
Classical filtering methods such as mean, median, and Gaussian blur smooth or denoise images by averaging or weighting pixel values in a local neighborhood. These filters reduce noise but may blur sharp edges and fine details, which can be a drawback if edge preservation is important.

**Thresholding:**
Thresholding techniques convert grayscale images into binary images by selecting a cutoff pixel intensity. Pixels with intensity above the threshold are set to one value (e.g., white), and pixels below are set to another (e.g., black). This binary segmentation is useful for separating foreground objects from the background.

**Adaptive Gaussian Thresholding:**
Adaptive Gaussian thresholding calculates the threshold for each pixel based on the weighted sum of neighboring pixel intensities using a Gaussian window. This local adaptation allows the threshold to dynamically adjust to different areas of the image, making it highly effective for images with uneven illumination or varying lighting conditions.

With adaptive Gaussian thresholding:

Foreground and background segmentation is more accurate in varying light regions.

* Foreground and background segmentation is more accurate in varying light    regions.
* Noise and shadows in the background are minimized.
* Edges and details in foreground objects are better preserved compared to global thresholding.










```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
```

Get all JPG images in the current directory

```python
image_files = glob.glob("*.jpg")
```

Create an output folder

```python
os.makedirs("thresholded_images", exist_ok=True)
```

Loop through each image and apply Adaptive Gaussian Threshold

```python
for img_file in image_files:
    # Read image in grayscale
    img = cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)

    if img is None:
        print(f"⚠️ Could not read {img_file}")
        continue
```

Apply Adaptive Gaussian Threshold

```python
th_img = cv2.adaptiveThreshold(img, 255,
                                   cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                   cv2.THRESH_BINARY,
                                   11, 2)
```

Save the thresholded image

```python
out_path = f"thresholded_images/threshold_{img_file}"
cv2.imwrite(out_path, th_img)
```

Display original and processed

```python
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(th_img, cmap='gray')
plt.title("Gaussian Mean Threshold")
plt.axis('off')

plt.suptitle(f"Processed: {img_file}")
plt.show()

print("✅ All images processed and saved in 'thresholded_images' folder.")
```

# **Analysis & Interpretation**

The adaptive Gaussian thresholding technique successfully segmented the butterfly from the background, even with the presence of lighting variations and background textures. Unlike global thresholding, which applies a single cutoff value across the image, the adaptive method adjusted thresholds locally based on pixel neighborhoods, preserving finer wing patterns and edges.

However, the method also captured some background noise in the form of scattered black speckles, which could be minimized by applying a pre-filtering step such as Gaussian blur or median filtering before thresholding. Overall, the adaptive Gaussian approach demonstrated strong edge preservation and accurate foreground separation, making it well-suited for images with uneven illumination, as in this butterfly example.
